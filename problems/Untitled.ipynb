```python
from glob import glob 
import pandas as pd

len_prefix = len("MCF7_")
len_suffix = len("_narrowPeak.bed")

dataframes = []

archivos = glob("MCF7*")
for file in archivos:
    f = file[len_prefix:-len_suffix]
    data = pd.read_csv(file, sep='\t',header=None)
    cromosomas = data[0].values
    peaks = data[9].values
    origin = [f for i in range(peaks.size)]
    dataframes.append(pd.DataFrame({
        'cromosoma' : cromosomas,
        'origen': origin,
        'peaks': peaks
    }))

pd.concat(dataframes).set_index(['cromosoma','origen']).sort_index()
```
